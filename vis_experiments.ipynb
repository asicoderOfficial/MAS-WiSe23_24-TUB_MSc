```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
import os

proportions = {}
for dir in os.listdir('experiments/10 agents/'):
    end_number = int(dir.split('-')[0])
    print(end_number)
    proportions[end_number] = pd.read_csv('experiments/10 agents/' + dir + '/delivery_data.csv')
```

```python
import numpy as np
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300

x = []
y = []
for key in sorted(proportions.keys()):
    x.append(key)
    y.append(proportions[key]["Delivery Time"].median())
    print(key, len(proportions[key]["PackagePoint X"]))

plt.figure(figsize=(7, 5))
plt.plot(x, y, marker='o')
plt.title('Median Delivery Time in 1000 Iterations')
plt.ylabel('Median Delivery Time (iterations)')
plt.xlabel('Agents with END goal/Total agents (percents)')
plt.savefig("experiment12.png")
print(x)
print(y)
```

```python


with_results = {
    "GRID1": pd.read_csv("experiments/with_n_without/with/GRID1-2023-11-24 19:59:15.280855/delivery_data.csv"),
    "GRID2": pd.read_csv("experiments/with_n_without/with/GRID2-2023-11-24 19:57:01.022562/delivery_data.csv"),
    "GRID3": pd.read_csv("experiments/with_n_without/with/GRID3-2023-11-24 19:41:34.829486/delivery_data.csv")
}

without_results = {
    "GRID1": pd.read_csv("experiments/with_n_without/without/GRID1-2023-11-24 20:02:16.774975/delivery_data.csv"),
    "GRID2": pd.read_csv("experiments/with_n_without/without/GRID2-2023-11-24 20:03:43.652130/delivery_data.csv"),
    "GRID3": pd.read_csv("experiments/with_n_without/without/GRID3-2023-11-24 20:06:06.670136/delivery_data.csv")
}

with_results_len = [len(v) for k, v in sorted(with_results.items())]
without_results_len = {k: len(v) for k, v in sorted(without_results.items())}   
print(sorted(without_results.keys()))

grids = ["GRID1", "GRID2", "GRID3"]
plot_data = {
    "With IPs (Intermediate 0.8, End 0.2)": (with_results["GRID1"]["Delivery Time"].median(), with_results["GRID2"]["Delivery Time"].median(), with_results["GRID3"]["Delivery Time"].median()),
    "Without IPs": (without_results["GRID1"]["Delivery Time"].median(), without_results["GRID2"]["Delivery Time"].median(), without_results["GRID3"]["Delivery Time"].median())
}

x = np.arange(len(grids))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

plt.figure(figsize=(5, 3))
fig, ax = plt.subplots(layout='constrained')
for attribute, measurement in plot_data.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

ax.set_xticks(x + width, grids)
ax.legend(loc='lower right', ncols=1)
ax.set_xlabel('Grids')
ax.set_ylabel('Median Delivery Time (iterations)')
ax.set_title("Median Delivery Time in 1000 Iterations")
plt.savefig("experiment22.png")
```
